# ðŸŽ¬ Movie Recommendation System

A content-based movie recommendation system using TF-IDF vectorization, K-Means clustering, and cosine similarity.

## Features
- **Content-Based Filtering**: Recommends movies based on genres, cast, and director
- **Hybrid Recommendations**: Combines content similarity with popularity scores
- **K-Means Clustering**: Groups similar movies for efficient recommendations
- **Fuzzy Search**: Finds movies even with typos in the search query

---

## 1. Setup and Imports

Import all required libraries for data processing, visualization, and machine learning.

```python
# Core Data Processing
import pandas as pd
import numpy as np
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Machine Learning
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Text Processing
from difflib import get_close_matches

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 50)
```

## 2. Data Loading & Initial Exploration

Load the movies dataset and perform initial exploration to understand its structure.

```python
# Robust data path resolution
def find_data_file():
    """Find the movies.csv file in common locations."""
    possible_paths = [
        Path('../data/movies.csv'),
        Path('./data/movies.csv'),
        Path('movies.csv'),
        Path('/kaggle/input/tmdb-movie-metadata/movies.csv'),
    ]
    
    for path in possible_paths:
        if path.exists():
            return path
    
    raise FileNotFoundError('movies.csv not found. Please ensure the data file is in the data/ directory.')

# Load the dataset
data_path = find_data_file()
df_raw = pd.read_csv(data_path)
print(f'Dataset loaded: {df_raw.shape[0]:,} movies, {df_raw.shape[1]} features')
```

```python
# Display first few rows
df_raw.head()
```

```python
# Dataset info
print('Dataset Information:')
print('-' * 50)
print(f'Total Movies: {len(df_raw):,}')
print(f'Total Features: {len(df_raw.columns)}')
print(f'Memory Usage: {df_raw.memory_usage(deep=True).sum() / 1024**2:.2f} MB')
print('\nColumn Types:')
df_raw.dtypes
```

## 3. Data Preprocessing & Feature Engineering

Clean the data by handling missing values, extracting relevant features, and creating new derived features.

```python
# Check for duplicates
duplicates = df_raw.duplicated().sum()
print(f'Duplicate rows: {duplicates}')

# Check for missing values
print('\nMissing values per column:')
missing = df_raw.isnull().sum()
missing_pct = (missing / len(df_raw) * 100).round(2)
missing_df = pd.DataFrame({'Missing': missing, 'Percentage': missing_pct})
missing_df[missing_df['Missing'] > 0].sort_values('Percentage', ascending=False)
```

```python
# Create a copy for processing
df = df_raw.copy()

# Drop columns with high missing values (homepage, keywords, tagline)
cols_to_drop = ['homepage', 'keywords', 'tagline', 'index']
cols_existing = [c for c in cols_to_drop if c in df.columns]
df = df.drop(columns=cols_existing)
print(f'Dropped columns: {cols_existing}')

# Drop rows with remaining missing values
rows_before = len(df)
df = df.dropna()
rows_after = len(df)
print(f'Removed {rows_before - rows_after} rows with missing values')
print(f'Remaining records: {rows_after:,}')
```

```python
# Extract release year from release_date
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['release_year'] = df['release_date'].dt.year.fillna(0).astype(int)

# Create decade column
df['decade'] = (df['release_year'] // 10 * 10).astype(str) + 's'

# Calculate profit and ROI
df['profit'] = df['revenue'] - df['budget']
df['roi'] = np.where(df['budget'] > 0, (df['profit'] / df['budget']) * 100, 0)

# Categorize profitability
def categorize_profit(profit):
    if profit > 100_000_000:
        return 'Blockbuster'
    elif profit > 0:
        return 'Profitable'
    elif profit == 0:
        return 'Break-even'
    else:
        return 'Loss'

df['profit_category'] = df['profit'].apply(categorize_profit)

print('Feature engineering complete!')
print(f'New features: release_year, decade, profit, roi, profit_category')
```

```python
# Select relevant columns for the recommendation system
feature_cols = ['original_title', 'original_language', 'genres', 'director', 'cast',
                'runtime', 'release_year', 'decade', 'budget', 'revenue', 'profit', 
                'roi', 'profit_category', 'vote_average', 'vote_count']

# Keep only existing columns
feature_cols = [c for c in feature_cols if c in df.columns]
df = df[feature_cols].copy()

# Reset index
df.reset_index(drop=True, inplace=True)

print(f'Final dataset shape: {df.shape}')
df.head()
```

## 5. Model Building: TF-IDF Vectorization & K-Means Clustering

Build the content-based filtering model using TF-IDF vectors and K-Means clustering.

```python
# Create movie profile by combining text features
df['profile'] = df['genres'].fillna('') + ' ' + df['cast'].fillna('') + ' ' + df['director'].fillna('')

# TF-IDF Vectorization
print('Vectorizing movie profiles...')
tfidf = TfidfVectorizer(stop_words='english', max_features=5000)
tfidf_matrix = tfidf.fit_transform(df['profile'])

print(f'TF-IDF Matrix Shape: {tfidf_matrix.shape}')
print(f'   - {tfidf_matrix.shape[0]:,} movies')
print(f'   - {tfidf_matrix.shape[1]:,} features (unique terms)')
```

```python
# K-Means Clustering
n_clusters = 25
print(f'Performing K-Means clustering with {n_clusters} clusters...')

kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(tfidf_matrix)

print(f'Clustering complete!')

# Cluster distribution
cluster_dist = df['cluster'].value_counts().sort_index()
print(f'\nCluster Size Statistics:')
print(f'   - Min cluster size: {cluster_dist.min()}')
print(f'   - Max cluster size: {cluster_dist.max()}')
print(f'   - Mean cluster size: {cluster_dist.mean():.0f}')
```

```python
# Visualize cluster distribution
fig = px.bar(x=cluster_dist.index, y=cluster_dist.values,
             title='Movies Per Cluster',
             labels={'x': 'Cluster ID', 'y': 'Number of Movies'},
             template='plotly_white',
             color=cluster_dist.values,
             color_continuous_scale='Turbo')
fig.update_layout(showlegend=False)
fig.show()
```

```python
# PCA for cluster visualization
print('Performing PCA for visualization...')
pca = PCA(n_components=2, random_state=42)
tfidf_2d = pca.fit_transform(tfidf_matrix.toarray())

df['pca_x'] = tfidf_2d[:, 0]
df['pca_y'] = tfidf_2d[:, 1]

print(f'Explained variance ratio: {sum(pca.explained_variance_ratio_)*100:.1f}%')

# Plot clusters
fig = px.scatter(df, x='pca_x', y='pca_y', color='cluster',
                 hover_data=['original_title', 'genres'],
                 title='Movie Clusters (PCA Visualization)',
                 labels={'pca_x': 'PCA Component 1', 'pca_y': 'PCA Component 2'},
                 template='plotly_white',
                 color_continuous_scale='Turbo',
                 opacity=0.6)
fig.update_layout(showlegend=True)
fig.show()
```

## 6. Recommendation Engine

Build the recommendation system with multiple features:
- Content-based recommendations using cosine similarity
- Fuzzy search for movie names
- Hybrid recommendations combining content and popularity
- Genre and year filtering options

```python
class MovieRecommender:
    """
    A content-based movie recommendation system.
    
    Features:
    - Fuzzy search for movie titles
    - Content-based filtering using TF-IDF and cosine similarity
    - Hybrid recommendations with popularity weighting
    - Genre and year filtering
    """
    
    def __init__(self, df, tfidf_matrix):
        self.df = df
        self.tfidf_matrix = tfidf_matrix
        self.movie_titles = df['original_title'].str.lower().tolist()
        
    def find_movie(self, title, threshold=0.6):
        """
        Find a movie by title using fuzzy matching.
        
        Args:
            title: Movie title to search for
            threshold: Minimum similarity threshold (0-1)
            
        Returns:
            Tuple of (found_title, index) or (None, None) if not found
        """
        title_lower = title.lower().strip()
        
        # Exact match first
        if title_lower in self.movie_titles:
            idx = self.movie_titles.index(title_lower)
            return self.df.iloc[idx]['original_title'], idx
        
        # Fuzzy match
        matches = get_close_matches(title_lower, self.movie_titles, n=1, cutoff=threshold)
        if matches:
            idx = self.movie_titles.index(matches[0])
            return self.df.iloc[idx]['original_title'], idx
        
        return None, None
    
    def get_recommendations(self, title, top_n=10, genre_filter=None, 
                            year_range=None, hybrid=False, content_weight=0.7):
        """
        Get movie recommendations based on a given movie.
        
        Args:
            title: Movie title to get recommendations for
            top_n: Number of recommendations to return
            genre_filter: Optional genre to filter by (e.g., 'Action')
            year_range: Optional tuple of (min_year, max_year)
            hybrid: Use hybrid scoring (content + popularity)
            content_weight: Weight for content similarity in hybrid mode (0-1)
            
        Returns:
            DataFrame with recommendations and similarity scores
        """
        # Find the movie
        found_title, idx = self.find_movie(title)
        
        if found_title is None:
            # Suggest similar titles
            suggestions = get_close_matches(title.lower(), self.movie_titles, n=5, cutoff=0.3)
            if suggestions:
                suggestion_titles = [self.df[self.df['original_title'].str.lower() == s]['original_title'].values[0] for s in suggestions]
                return f"Movie '{title}' not found. Did you mean: {', '.join(suggestion_titles)}?"
            return f"Movie '{title}' not found in database."
        
        print(f"ðŸŽ¬ Finding recommendations for: '{found_title}'")
        
        # Get cluster
        movie_cluster = self.df.loc[idx, 'cluster']
        cluster_movies = self.df[self.df['cluster'] == movie_cluster].copy()
        
        # Calculate cosine similarity within cluster
        cosine_sim = cosine_similarity(
            self.tfidf_matrix[idx], 
            self.tfidf_matrix[cluster_movies.index]
        ).flatten()
        
        cluster_movies['similarity'] = cosine_sim
        
        # Hybrid scoring
        if hybrid:
            # Normalize popularity (vote_count * vote_average)
            max_pop = (cluster_movies['vote_count'] * cluster_movies['vote_average']).max()
            cluster_movies['popularity_score'] = (cluster_movies['vote_count'] * cluster_movies['vote_average']) / max_pop
            cluster_movies['hybrid_score'] = (content_weight * cluster_movies['similarity'] + 
                                              (1 - content_weight) * cluster_movies['popularity_score'])
            sort_col = 'hybrid_score'
        else:
            sort_col = 'similarity'
        
        # Apply filters
        if genre_filter:
            cluster_movies = cluster_movies[cluster_movies['genres'].str.contains(genre_filter, case=False, na=False)]
        
        if year_range:
            cluster_movies = cluster_movies[
                (cluster_movies['release_year'] >= year_range[0]) & 
                (cluster_movies['release_year'] <= year_range[1])
            ]
        
        # Sort and get top recommendations (excluding the input movie)
        recommendations = cluster_movies[cluster_movies.index != idx].nlargest(top_n, sort_col)
        
        # Format output
        output_cols = ['original_title', 'genres', 'director', 'release_year', 
                       'vote_average', 'similarity']
        if hybrid:
            output_cols.append('hybrid_score')
        
        return recommendations[output_cols].round(3)
    
    def explain_recommendation(self, source_movie, target_movie):
        """
        Explain why a movie was recommended.
        
        Args:
            source_movie: Original movie
            target_movie: Recommended movie
            
        Returns:
            Dictionary with explanation
        """
        source_title, source_idx = self.find_movie(source_movie)
        target_title, target_idx = self.find_movie(target_movie)
        
        if source_title is None or target_title is None:
            return "One or both movies not found."
        
        source_data = self.df.iloc[source_idx]
        target_data = self.df.iloc[target_idx]
        
        # Calculate similarity
        sim = cosine_similarity(self.tfidf_matrix[source_idx], self.tfidf_matrix[target_idx])[0][0]
        
        return {
            'source_movie': source_title,
            'recommended_movie': target_title,
            'similarity_score': round(sim, 3),
            'shared_genres': set(source_data['genres'].split()) & set(target_data['genres'].split()),
            'same_director': source_data['director'] == target_data['director'],
            'same_cluster': source_data['cluster'] == target_data['cluster']
        }

# Initialize the recommender
recommender = MovieRecommender(df, tfidf_matrix)
print('Movie Recommender initialized!')
```
